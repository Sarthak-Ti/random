```python
#we will randomize the figure assignments

figures = [1,2,3,4]
students = ['karthik','feiyang','jon','austin','cheryl','sherry','mythili','berk','hoyin','afsana','david','kasey','harry','robby','taykhoom','sam']
print(len(students))
assignments = {}
#now randomly shuffle students
import random
random.seed(a=420)
random.shuffle(students)
for i,student in enumerate(students):
    assignments[student] = figures[i%4]
print(assignments)
```

```python
import pandas as pd

students_df = pd.DataFrame(list(assignments.items()), columns=['Name', 'Figure'])

# Now we'll sort the DataFrame based on the 'Figure' column to group them
students_df_sorted = students_df.sort_values(by='Figure')

students_df_sorted.reset_index(drop=True, inplace=True)
students_df_sorted
```
